#### 凯撒加密，每个字母替换为后面第k个
- ascii_letters：包含所有 ASCII 字母的字符串，即包括小写字母和大写字母。
  它的值为 'abcdefghijklmnopqrstuvwxyzABCDEFGHIJKLMNOPQRSTUVWXYZ'。

- ascii_lowercase：包含所有小写 ASCII 字母的字符串。它的值为 'abcdefghijklmnopqrstuvwxyz'。

- ascii_uppercase：包含所有大写 ASCII 字母的字符串。它的值为 'ABCDEFGHIJKLMNOPQRSTUVWXYZ'。

```python
from string import ascii_letters, ascii_lowercase, ascii_uppercase

def kaisaEncrypt(text, k):
    # 凯撒加密
    lower = ascii_lowercase[k:]+ascii_lowercase[:k] #根据给定的偏移量 k，生成一个新的小写字母表
    upper = ascii_uppercase[k:]+ascii_uppercase[:k] #根据给定的偏移量 k，生成一个新的大写字母表
    table = ''.maketrans(ascii_letters, lower+upper) #创建映射表
    return text.translate(table) #按映射表进行替换

s = "Python is a greate programming language. I like it!"
st=kaisaEncrypt(s, 3)
print(st)
```

#### 查找一个字符串中最长的数字子串

```python
def longgest(s):
    # 返回字符串中最长的单词
    for i in s:
        if i.isdigit()==False:
            s=s.replace(i," ")
    return max(s.split(), key=len)
longgest("3879ije ieei7865765673jkhg787ybh6583kj")
```
