#### random.randrange(start, stop[, step])：生成一个指定范围内的随机整数

### itertools模块

#### itertools.permutations(iterator, r)：重在排列,接受一个迭代器和需要排列的长度r作为输入，并假设“r”作为迭代器的默认长度，并分别返回所有可能的长度为r的排列组合。

```python
from itertools import  permutations
a="abc"
print (permutations(a,3))
print (type(permutations(a,3)))
for i in permutations(a,3):#返回字符串a中元素所有可能的排列组合
    x = ''.join(i)
    print (x,end=' ')
```

```python
from itertools import permutations
def isPrime(n):
    for i in range(2,int(pow(n,0.5))+1):
        if n%i==0:
            return False
    return True

a=(1,2,3,4)
for i in permutations(a,4):
    n=int(''.join(map(str,i)))
    if isPrime(n):
        print(n)
```

#### itertools.combinations(lst, n):重在组合，返回一个迭代器，生成给定列表lst中长度为n的所有可能组合。
例如，如果lst为[1, 2, 3]，n为2，那么combinations(lst, n)将生成(1, 2)、(1, 3)和(2, 3)这三个组合。

```python
from itertools import combinations
# 字符串的组合
string = "abc"
combinations_list = list(combinations(string, 2))
print(combinations_list)
# 输出: [('a', 'b'), ('a', 'c'), ('b', 'c')]

# 列表的组合
lst = [1, 2, 3]
combinations_list = list(combinations(lst, 2))
print(combinations_list)
# 输出: [(1, 2), (1, 3), (2, 3)]

# 元组的组合
tuple = (1, 2, 3, 4)
combinations_list = list(combinations(tuple, 3))
print(combinations_list)
# 输出: [(1, 2, 3), (1, 2, 4), (1, 3, 4), (2, 3, 4)]
```

#### itertools.chain():接受一个可迭代对象列表作为输入，并返回一个迭代器，有效的屏蔽掉在多个容器中迭代细节。

```python
vec = [[1,2,3], [4,5,6], [7,8,9]]
a=[num for elem in vec for num in elem]
print(a)

#相当于
from itertools import chain
print(list(chain(*vec))) #取vec中的值
```

### copy模块
> **copy模块**的**deepcopy()**方法：
*   对原列表中的元素进行递归，把所有的值都复制到新列表中，对嵌套的子列表不再是复制引用。
*   此时新列表和原列表是互相独立，修改任何一个都不会影响另外一个。

```python
a = [3, [5], 7]
import copy
b = copy.deepcopy(a) #深复制，递归复制直到遇到可哈希对象
                #a和b完全独立，互相不影响
print(a == b) #a、b元素完全相同
print(a is b) #a、b元素不是一个对象
print(b[1].append(6))  #修改b不会影响a
print(b)
print(a)
```

### operator模块：提供了一系列用于进行对象操作的函数。
这些函数可以用于自定义类的操作符重载（operator overloading），以及对内置类型进行更高级别的操作

对于需要处理大量数据、进行函数式编程或简化代码逻辑的场景特别有用。

```python
import operator

# 算数计算
a = 5
b = 3
result = operator.add(a, b)# 加法
print(result)  # 输出: 8
result = operator.mul(a, b)# 乘法
print(result)  # 输出: 15
result = operator.truediv(a, b) # 除法
print(result)  # 输出: 1.6666666666666667
```

#### operator.itemgetter():获取对象的属性或索引

```python
person = {'name': 'Alice', 'age': 25, 'city': 'New York'}

get_age = operator.itemgetter('age')
age = get_age(person)
print(age)  # 输出: 25

get_name = operator.itemgetter('name')
name = get_name(person)
print(name)  # 输出: 'Alice'
```

#### operator.methodcaller():调用对象方法

```python
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def get_older(self, n):
        self.age += n   
        return self.age

person = Person('Alice', 25)
older_func = operator.methodcaller('get_older',5)
result = older_func(person)
print(result)  
```

## collections模块
内置模块，提供了一些额外的数据结构类型，用于扩展内置的数据类型（如列表、字典和元组）以及提供更多的操作和功能

### **Counter**对象：计数器，用于计数可哈希对象的出现次数。
它是一个字典的子类，提供了快速统计元素出现频率的功能。

```python
from collections import Counter

# 统计列表中元素的出现次数
fruits = ['apple', 'banana', 'apple', 'orange', 'banana', 'apple']
fruit_counts = Counter(fruits)
print(fruit_counts)
```

```python
cnt = Counter()
for word in ['red', 'blue', 'red', 'green', 'blue', 'blue']:
    cnt[word] += 1
print(cnt)
```

```python
import re
words = re.findall(r'\w+', open('article.txt').read().lower())
# re.findall() 方法返回一个包含所有匹配结果的列表
# \w+ 表示匹配由连续的字母、数字或下划线组成的字符串
Counter(words).most_common(10)      #统计列表 words 中每个单词的出现次数，返回出现次数最多的10个单词
```

#### most_common(n)方法：返回出现次数最多的 n 个元素和对应的出现次数的列表
按照出现次数从高到低的顺序排序

```python
fruit_counts.most_common(2) # 出现次数最多的两个元素   
```

#### elements()方法：返回一个迭代器，按照计数的顺序依次重复每个元素。

```python
list(fruit_counts.elements()) # 返回一个迭代器，按计数顺序包含所有出现的元素
```

#### update(iterable)方法：从另一个可迭代对象中更新计数。

```python
fruit_counts.update(['orange', 'orange', 'orange']) # 更新计数
print(fruit_counts)
```

#### subtract(iterable)方法：从另一个可迭代对象中减去计数。

```python
fruit_counts.subtract(['orange']) # 减少计数
print(fruit_counts)
```

### **OrderedDict**对象：有序字典，保留了字典中元素的插入顺序。
与普通字典不同，它可以按照插入顺序遍历键值对

### **defaultdict**对象：默认字典，是字典的一个子类，提供了一个默认值的机制。
当访问字典中不存在的键时，会返回一个默认值，而不是引发 KeyError 错误。

```python
from collections import defaultdict

d = defaultdict(int)# 创建一个默认值为0的 defaultdict 对象
# 使用 defaultdict 时需要提供一个默认值的类型，这里使用了 int 类型作为默认值，表示默认值为0
# 根据具体需求，也可以使用其他类型作为默认值。

fruits = ['apple', 'banana', 'apple', 'orange', 'banana', 'apple']
for fruit in fruits:
    d[fruit] += 1
print(d) #字典 d 中的键值对是每个水果和其对应的出现次数
```

## timeit模块

### timeit.Timer(stmt='pass', setup='pass', timer=<timer function>)：
- stmt：要测量执行时间的语句或代码块。可以是字符串形式的代码片段，或是多行代码的字符串。
- setup：可选参数，用于指定执行语句所需的上下文环境。一般情况下，使用默认值 'pass' 即可，除非需要导入模块或设置一些全局变量。
- timer：可选参数，用于指定计时器函数。默认使用系统计时器。

#### Timer.timeit(num) 方法:返回执行代码所用的总时间（以秒为单位）
num为可选参数，用于指定重复执行的次数。默认值是 1000000。

```python
import timeit # Python 内置模块，用于测量代码的执行时间

#使用列表推导式生成10000个字符串
strlist = ['This is a long string that will not keep in memory.' 
           for n in range(10000)]
def use_join():
    return ''.join(strlist)
def use_plus():
    result = ''
    for str in strlist:
        result = result+str
    return result
if __name__ == '__main__':
    times = 1000 #重复运行次数
    jointimer = timeit.Timer('use_join()', 'from __main__ import use_join') #创建 Timer 对象
    print('time for join:', jointimer.timeit(number=times)) #测量函数 use_join() 的执行时间
    plustimer = timeit.Timer('use_plus()', 'from __main__ import use_plus')
    print('time for plus:', plustimer.timeit(number=times))
```

### zlib模块
#### zlib.**compress**()和zlib.**decompress**():用于**字节串**的压缩和解压缩

```python
import zlib
x = 'Python程序设计'*5
x = x.encode() #编码

xp = zlib.compress(x) #压缩

xpd=zlib.decompress(xp) #解压
print(xpd.decode()) #解码

print(len(x),len(xp),len(xpd)) #信息重复度越高，压缩比越大
```
