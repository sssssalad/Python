## 数值处理
### type(a):显示变量x的数据类型

```python
a=1
print("a:",type(a))#整数
b=1.1
print("b:",type(b))#浮点数
c=True
print("c:",type(c))#布尔值
d=complex(1,2)
print("d:",type(d))#复数
e=1+2j
print("e:",type(e))#复数
```

```python
e.real#查看复数实部
```

```python
e.imag#查看复数虚部
```

```python
e.conjugate()#返回共轭复数
```

### isinstance():测试对象是否为指定类型

```python
x=3
isinstance(x,int)
```

## 浮点数float：(带小数点的数字）
### 浮点数加减结果包含的小数位数可能不确定

```python
print(0.1+0.1)
print(2*0.1)
print(0.2+0.1)
```

### 受浮点数精度影响,取余计算结果可能有误差

```python
print(3.1%2)
print(6.3%2.1)
print(5.7%4.8)
```
