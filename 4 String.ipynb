#### 性能比较

#### timeit.Timer(stmt='pass', setup='pass', timer=<timer function>)：
- stmt：要测量执行时间的语句或代码块。可以是字符串形式的代码片段，或是多行代码的字符串。
- setup：可选参数，用于指定执行语句所需的上下文环境。一般情况下，使用默认值 'pass' 即可，除非需要导入模块或设置一些全局变量。
- timer：可选参数，用于指定计时器函数。默认使用系统计时器。
#### Timer.timeit(num) 方法:返回执行代码所用的总时间（以秒为单位）
num为可选参数，用于指定重复执行的次数。默认值是 1000000。

```python
import timeit # Python 内置模块，用于测量代码的执行时间

#使用列表推导式生成10000个字符串
strlist = ['This is a long string that will not keep in memory.' 
           for n in range(10000)]
def use_join():
    return ''.join(strlist)
def use_plus():
    result = ''
    for str in strlist:
        result = result+str
    return result
if __name__ == '__main__':
    times = 1000 #重复运行次数
    jointimer = timeit.Timer('use_join()', 'from __main__ import use_join') #创建 Timer 对象
    print('time for join:', jointimer.timeit(number=times)) #测量函数 use_join() 的执行时间
    plustimer = timeit.Timer('use_plus()', 'from __main__ import use_plus')
    print('time for plus:', plustimer.timeit(number=times))
```

```python
timeit.timeit('"-".join(str(n) for n in range(100))', number=10000)  
#生成器表达式，每生成一个字符串就使用join方法，比列表推导式更节省内存
```

```python
timeit.timeit('"-".join([str(n) for n in range(100)])', number=10000) 
#列表推导式，生成了整个列表再使用join方法
```

```python
timeit.timeit('"-".join(map(str, range(100)))', number=10000) #使用map方法类型转换较高效
```

## 字符串格式化
将其他类型对象按要求转换为字符串并返回结果字符串
>'% [-] [+] [0] [m] [.n] 格式字符 '% x
    >> - '%：格式标志，表示格式开始
    >> - [-]：指定左对齐输出
    >> - [+]：对正数加正号
    >> - [0]：指定空位填0
    >> - [m]：指定**最小宽度**
    >> - [.n]：指定精度
    >> - **格式字符**：指定类型
    >> - '%：格式运算符 
    >> - x：待转换的表达式

格式字符：
> - %s：字符串，类似str()
> - %r：字符串，类似repr()

> - %g：指数(e)或浮点数(根据显示长度)
> - %G：指数(E)或浮点数(根据显示长度)
> - %%：一个字符%

#### 生成随机密码
random.choice(seq)：从序列seq中随机选择一个元素

```python
import string
import random
characters = string.digits + string.ascii_letters #数字+字母
print(''.join([random.choice(characters) for i in range(8)])) #生成8位随机密码
print(''.join([random.choice(characters) for i in range(10)])) #生成10位随机密码
print(''.join([random.choice(characters) for i in range(16)])) #生成16位随机密码
print(''.join(random.choices(characters, k=12))) #生成12位随机密码
```

#### 凯撒加密，每个字母替换为后面第k个
- ascii_letters：包含所有 ASCII 字母的字符串，即包括小写字母和大写字母。
  它的值为 'abcdefghijklmnopqrstuvwxyzABCDEFGHIJKLMNOPQRSTUVWXYZ'。

- ascii_lowercase：包含所有小写 ASCII 字母的字符串。它的值为 'abcdefghijklmnopqrstuvwxyz'。

- ascii_uppercase：包含所有大写 ASCII 字母的字符串。它的值为 'ABCDEFGHIJKLMNOPQRSTUVWXYZ'。

```python
from string import ascii_letters, ascii_lowercase, ascii_uppercase

def kaisaEncrypt(text, k):
    # 凯撒加密
    lower = ascii_lowercase[k:]+ascii_lowercase[:k] #根据给定的偏移量 k，生成一个新的小写字母表
    upper = ascii_uppercase[k:]+ascii_uppercase[:k] #根据给定的偏移量 k，生成一个新的大写字母表
    table = ''.maketrans(ascii_letters, lower+upper) #创建映射表
    return text.translate(table) #按映射表进行替换

s = "Python is a greate programming language. I like it!"
st=kaisaEncrypt(s, 3)
print(st)
```

#### 给定任意字符串，查找其中每个字符的最后一次出现，并按每个字符最后一次出现的先后顺序依次存入列表
例如对于字符串'abcda'的处理结果为['b', 'c', 'd', 'a']，而字符串'abcbda'的处理结果为['c', 'b', 'd', 'a']。

```python
#s=input('请输入一段文本：')
s="aaaabcdawerasdfasdfwerngsnnvAAAweB3a"
import time
```

```python
start=time.time()

d={i:s.rfind(i) for i in s} #rfind()返回字符串最后一次出现的位置，如果没有匹配项则返回-1
d=sorted(d.items(),key=lambda x:x[1]) #按照字典的值进行排序
#print(d)
#d.items()返回一个可遍历的(键, 值) 元组数组
print([key for key,value in d])

stop=time.time()
print('运行时间：',stop-start)
```

```python
import re
def rev3(s):
    '''考虑开头或结束有空格的情况'''
    t = re.split('\s+', s.strip()) #strip()去除字符串开头和结尾的空格
    t.reverse()
    return ' '.join(t)

def rev4(s):
    '''考虑开头或结束有空格的情况'''
    t = re.split('\s+', s.strip()) #strip()去除字符串开头和结尾的空格
    return ' '.join(reversed(t))
```

#### 查找一个字符串中最长的数字子串

```python
def longgest(s):
    # 返回字符串中最长的单词
    for i in s:
        if i.isdigit()==False:
            s=s.replace(i," ")
    return max(s.split(), key=len)
longgest("3879ije ieei7865765673jkhg787ybh6583kj")
```

### 有一些句子和一些关键词，现在想找出包含至少一个关键词的那些句子（文本嗅探）

```python
sentences = ['This is a test.',
             'Beautiful is better than ugly.',
             'Explicit is better than implicit.',
             'Simple is better than complex.',
             'Sparse is better than dense.',
             'Readability counts.',
             'Now is better than never.']
words = ['test', 'count', 'dense', 'is', 'simple']
```

```python
import time
```

```python
# 不能写成if ((w+" ") or (w+".")) in s.lower())
# 因为or的优先级>in,会先执行or返回True或False,再执行in
start=time.time()
for s in sentences:
    s_key=0
    key_w=[]
    for w in words:
        if ((w+" ") in s.lower()) or ((w+".") in s.lower()):
            s_key+=1
    if(s_key>=1):
        print(s,s_key/s.split().__len__())
stop=time.time()
print('运行时间：',stop-start)
```
