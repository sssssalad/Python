## str.count(sub，start=0，end=len(string))方法：统计字符串str里子字符串sub出现的次数

可选参数为在字符串搜索的开始与结束位置。

```python
str = "this is string example....wow!!!";
 
sub = "i";
#从第5-40个元素中查找sub，即‘i'字符
print("str.count(sub, 4, 40) : ", str.count(sub, 4, 40))

sub = "wow";
##从str查找子字符串isub，即‘wow'
print("str.count(sub) : ", str.count(sub))
```

```python
#统计一个子字符串中所有字符在另一个字符串中出现的总次数
#子字符串为s2
def check(s1, s2):
    return sum(map(s1.count, s2))

print(check('这是一个测*#试邮#件，内】含广【告', '【】*#/\\'))
```

## str.split(str="", num=string.count(str))：通过指定分隔符对字符串进行切片
如果参数 num 有指定值，则分隔 num+1 个子字符串

```python
s="hello world"
s=s.split()
for i in s:
  print(i,end='---')
```

```python
s="hello##world"
s=s.split()#默认分割空格""
for i in s:
  print(i,end='---')
```

```python
s="hello##world"
s=s.split('##')#默认分割空格""
for i in s:
  print(i,end='---')
```

## zip():可迭代的对象作为参数，将对象中对应的元素打包成一个个元组，然后返回由这些元组组成的列表。

### map(,data):对对象进行指定的转化，map(int,a)将a转化为int类型，map(metric,a)将metric函数作用于a

```python
s5=[]
#zip(s1,s2)将s1和s2中对应的元素打包成一个个元组
#map(list,zip(s1,s2))将zip(s1,s2)中的元组转换成列表形式
#list(map(list,zip(s1,s2)))：打印s1和s2中对应的元素最终打包成的每个小列表
s5=list(map(list,zip(s1,s2)))
s5
```
