### 向文本文件写入内容
- **open()**:接受一个参数:要打开的文件的名称；python在当前执行的文件所在的目录查找指定的文件，并返回一个表示文件的对象
- **write(s)**：将字符串s写入文件
- **close()**：把缓存区内容写入文件，同时关闭文件，并释放文件对象

文件打开模式：
- 'r' : 读模式
- 'w' : 写模式
- 'a' ：追加模式
- 'b' ：二进制模式，可与其他模式组合使用
- '+' ：读、写模式，可与其他模式组合使用

```python
f=open ('sample.txt','a+')
s='文本文件的读取方法\n 文本文件的写入方法\n '
f.write(s) #将字符串s写入文件
f.close() #把缓存区内容写入文件，同时关闭文件，并释放文件对象
```

```python
#更优雅的写法
s='文本文件的读取方法\n 文本文件的写入方法\n'
with open('sample.txt','a+') as f: f.write(s)
```

### 打开并读取文件

- **with**关键字:上下文管理关键字可以自动管理资源，在不再需要访问文件后将其关闭
- **read()**:读取文件中指定数量的字符,默认为全部内容，并将其作为一个长字符串存储在contents变量中，到达文件末尾时返回一个空字符串，显示出来就是一个空行（对中英文同等对待）

```python
with open('pi_digits.txt') as file_object:
    """在本py文件所在的目录中查找pi_digits.txt文件"""
    contents = file_object.read()
    print(contents)
```

#### **seek**(offset[, whence])：把文件指针移动到新的位置
- *offset*：表示相对于 whence 的位置
- *whence*：为0表示从文件头开始计算，为1表示从当前位置开始计算，2表示从文件尾开始计算，默认为 0

```python
with open('sample.txt','r') as f: 
    print(f.read(5)) # 读取前5个字符
    print(f.read(7)) # 包含两个换行符
    f.seek(0) # 将文件指针移动到文件开头
    print(f.read(8))
```

### 逐行读取
- 每行末尾都有一个看不见的换行符，print语句也会加上一个换行符
- 每行输出都有两个换行符，一个来自文件一个来自print

```python
filename = 'pi_digits.txt'
"""变量filename表示的不是实际文件,
    它是一个表示文件及其内容的对象,
    是一个让python知道到哪里去查找文件的字符串"""

with open(filename) as file_object:
    """通过对文件对象执行循环来遍历文件中的每一行"""
    for line in file_object:
        print(line)
```

### 使用try-expect代码块

#### try代码块代码运行无问题则跳过except代码块，有问题则查找except代码块并执行

#### 将可能引发错误的代码放到try-expect模块可提高程序抵御错误的能力

```python
try:
    print(5/0)#引发了ZeroDivisionError错误
except ZeroDivisionError:
    """告诉python当遇到ZeroDivisionError错误时怎么处理"""
    print("You can't divide by zero!")
```

```python
print("Give me two numbers, and I'll divide them.")
print("Enter 'q' to quit.")

while True:
    first_number = input("\nFirst number: ")
    if first_number == 'q':
        break
        
    second_number = input("Second number: ")
    if second_number == 'q':
        break  
    try:
        answer = int(first_number) / int(second_number)
    except ZeroDivisionError:
        print("You can't divide by 0!")
        #告诉用户为什么错误，避免显示traceback
        
    else:
        print(answer)
```

```python
def count_words(filename):
    """计算一个文件大致包含多少个单词"""
    try:
        with open(filename,encoding='utf-8') as f:
            contents = f.read()
            
    except FileNotFoundError:
        msg = "Sorry,the file ‘" + filename + "’ does not exist."
        print(msg)
    except UnicodeDecodeError:
        #代码太长时分行可使用括号
        msg =( "Sorry,the file ‘" + filename + 
              "’ can't decode.Please add （encoding='utf-8'） in open function.")
        print(msg)
        
    else:
        words = contents.split()#将文章内容生成一个列表
        num_words = len(words)
        print("The file ‘" + filename + "’ has about " + str(num_words) + " words.")

filename = "article.txt"
count_words(filename)
```

```python
import json

def get_stored_username():
    """如果存储了用户名，就获取它"""
    filename = 'username.json'
    try:
        with open(filename) as f:
            username = json.load(f)
    except FileNotFoundError:
        return None
    else:
        return username
    
def get_new_username():
    """提示用户输入用户名"""
    username = input("What is your name? ")
    filename = 'username.json'
    with open(filename,'w') as f:
        json.dump(username,f)
    return username
    
def greet_users():
    """先询问用户名是否正确，再问候用户并指出其名字"""
    username = get_stored_username()
    if username:
        msg = input("If the username: " + username + " is correct? yes or no")
        if msg == 'yes':
            print("Welcome back, " + username.title() + "!")
        else:
            username = get_new_username()
            print("Welcome back, " + username.title() + "!")
    else:
        username = get_new_username()
        print("We'll remember you when you come back, " + username.title() + "!")
        
greet_users()
```
