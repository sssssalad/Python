按**文件中数据的组织形式**把文件分为文本文件和二进制文件两类。
1. **文本文件**：
   - 文本文件存储的是**常规字符串**，由若干文本行组成，通常每行以换行符'\n'结尾。
   - 常规字符串是指记事本或其他文本编辑器能正常显示、编辑并且**人类能够直接阅读和理解**的字符串。
   - 如英文字母、汉字、数字字符串。
   - 文本文件可以使用字处理软件如gedit、记事本进行编辑。
2. **二进制文件**：
   - 二进制文件把对象内容以**字节串**(bytes)进行存储
   - 无法用记事本或其他普通字处理软件直接进行编辑，通常也**无法被人类直接阅读和理解**，需要使用专门的软件进行解码后读取、显示、修改或执行。
   - 常见的如图形图像文件、音视频文件、可执行文件、资源文件、各种数据库文件、各类office文档等都属于二进制文件。

文件内容基本操作：**打开**、**读写**、**关闭**

### 向文本文件写入内容
**open**(**file**, **mode**='r', buffering=-1, encoding=None, errors=None,newline=None, closefd=True, opener=None):
python在当前执行的文件所在的目录查找指定的文件，并返回一个表示文件的对象
- file:要打开的文件的名称；
- mode：指定打开文件后的打开模式（处理方式）
  - 'r' : 读模式,默认，可省略，文件不存在时报错
  - 'w' : 写模式，文件已存在时会清空原内容
  - 'x' : 写模式，创建新文件，文件已存在时报错
  - 'a' ：追加模式
  - 'b' ：二进制模式，可与其他模式组合使用
  - 't' ：文本模式，默认，可省略
  - '+' ：读、写模式，可与其他模式组合使用 
- buffering：指定读写文件的缓存模式。
  - 0表示不缓存，1表示缓存
  - 大于1则表示缓冲区的大小
  - 默认值-1表示由系统管理缓存。
- encoding：指定对文本进行编码和解码的方式，只适用于文本模式，可以使用Python支持的任何格式，如GBK、utf8、CP936等等。

- **write(s)**：将字符串s写入文件
- **close()**：把缓存区内容写入文件，同时关闭文件，并释放文件对象
- 当对文件内容操作完以后，一定要关闭文件对象，这样才能保证所做的任何修改都确实被保存到文件中：f.close()，但当关闭之前发生错误时无法保证文件一定能够正常关闭，推荐使用with关键字。

### 使用try-expect代码块

#### try代码块代码运行无问题则跳过except代码块，有问题则查找except代码块并执行

#### 将可能引发错误的代码放到try-expect模块可提高程序抵御错误的能力

```python
try:
    print(5/0)#引发了ZeroDivisionError错误
except ZeroDivisionError:
    """告诉python当遇到ZeroDivisionError错误时怎么处理"""
    print("You can't divide by zero!")
```

```python
print("Give me two numbers, and I'll divide them.")
print("Enter 'q' to quit.")

while True:
    first_number = input("\nFirst number: ")
    if first_number == 'q':
        break
        
    second_number = input("Second number: ")
    if second_number == 'q':
        break  
    try:
        answer = int(first_number) / int(second_number)
    except ZeroDivisionError:
        print("You can't divide by 0!")
        #告诉用户为什么错误，避免显示traceback
        
    else:
        print(answer)
```
